```python
# Carregando o arquivo

df = pd.read_csv("StudentsPerformance.csv")
```

```python
# Imprimindo as 5 primeiras linhas

print(df.head())
```

```python
# Informações gerais sobre o dataset

print(df.info())
```

```python
# Estatística descritiva do dataset

print(df.describe())
```

Início da análise exploratória

```python
# Distribuição de notas de matemática

plt.figure(figsize=(8, 6))
sns.histplot(df['math score'], bins=20, kde=True, color='blue')
plt.title('Distribuição das notas de mátematica')
plt.xlabel('Nota')
plt.ylabel('Frequência')
plt.show()
```

```python
# Notas médias por gênero

plt.figure(figsize=(8, 6))
sns.barplot(x='gender', y='math score', data=df, hue='gender', dodge=False, palette='pastel', legend=False)
plt.title('Notas Médias de Matemática por Gênero')
plt.xlabel('Gênero')
plt.ylabel('Nota média')
plt.show()
```

```python
# Insights: Diferença de notas entre alunos que fizeram ou não curso preparatório

plt.figure(figsize=(8, 6))
sns.boxplot(x='test preparation course', y='math score', data=df,
            hue='test preparation course', dodge=False, palette='muted', legend=False)
plt.title("Notas de matemática por Curso Preparatório")
plt.xlabel('Curso Preparatório')
plt.ylabel('Nota')
plt.show()
```

Calculando as médias das notas por cada nível de escolaridade dos pais

```python

medias_notas = df.groupby('parental level of education')[['math score', 'reading score', 'writing score']].mean()
```

Visualização gráfica

```python
# Nota de Mátematica por nível de escolaridade dos pais
plt.figure(figsize=(8, 6))
sns.barplot(x='parental level of education', y='math score', data=df, hue='parental level of education',
            dodge=False, palette='Set3', legend=False)
plt.title('Nota Média de Matématica por Nível de Escolaridade dos Pais.')
plt.xlabel('Nível de Escolaridade dos Pais')
plt.ylabel('Nota Média de Matemática')
plt.xticks(rotation=45)
plt.show()
```

```python
# Notas de leitura por nível de escolaridade dos pais
plt.figure(figsize=(8, 6))
sns.barplot(x='parental level of education', y='reading score', data=df, hue='parental level of education',
            dodge=False, palette='Set3', legend=False)
plt.title('Nota Média de Leitura por Nível de Escolaridade dos Pais')
plt.xlabel('Nível de escolaridade dos Pais')
plt.ylabel('Nota Média de Leitura')
plt.xticks(rotation=45)
plt.show()
```

```python
# Notas de escrita por nível de escolaridade dos Pais

plt.figure(figsize=(8, 6))
sns.barplot(x='parental level of education', y='writing score', hue='parental level of education', data=df,
            dodge=False, palette='Set3', legend=False)
plt.title('Nota Média de Escrita por nível de Escolaridade dos Pais')
plt.xlabel('Nível de escolaridade dos Pais')
plt.ylabel('Nota Média de Escrita')
plt.xticks(rotation=45)
plt.show()
```

Boxplots para verificar a dispersão das notas

```python
# Nível de Matemática

plt.figure(figsize=(8,6))
sns.boxplot(x='parental level of education', y='math score', data=df,
            hue='parental level of education', dodge=False,
            palette='Set3', legend=False)
plt.title('Dispersão de Notas de Leitura por Nível de Escolaridade dos Pais')
plt.xlabel('Nível de Escolaridade dos Pais')
plt.ylabel('Nota de Matemática')
plt.xticks(rotation=45)
plt.show()
```

```python
#Nível de Leitura

plt.figure(figsize=(8,6))
sns.boxplot(x='parental level of education', y='reading score', data=df,
            hue='parental level of education', dodge=False,
            palette='Set3', legend=False)
plt.title('Dispersão das Notas de Escrita por Nível de Escolaridade dos Pais')
plt.xlabel('Nível de escolaridade dos Pais')
plt.ylabel('Nota de Leitura')
plt.xticks(rotation=45)
plt.show()
```

```python
#Nível de Escrita

plt.figure(figsize=(8, 6))
sns.boxplot(x='parental level of education', y='writing score', data=df,
            hue='parental level of education', dodge=False, palette='Set3',
            legend=False)
plt.title('Dispersão de notas de Escrita por Nível de Escolaridade dos Pais')
plt.xlabel('Nível de Escolaridade dos Pais')
plt.ylabel('Notas de Escrita')
plt.xticks(rotation=45)
plt.show()
```
